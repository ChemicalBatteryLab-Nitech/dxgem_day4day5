## 1. Tensor形式で入力データ作成
pytorchでは、numpyのarray形式ではなく、独自の Tensor形式でデータをハンドリングします。<BR>
ただし、ほぼnumpyと同じような使い勝手です。<BR>

```python
Xnp=np.array([[0,0],[0,1],[1,0],[1,1]])
Xnp
```

```python
X=torch.Tensor([[0,0],[0,1],[1,0],[1,1]])
t=torch.Tensor([[0],[1],[1],[1]])
print (X)
print (t)
```

```python
#操作不要 (numpy-array > torch-tensor)   

Xtc = torch.from_numpy(Xnp)
Xtc
```

```python
#操作不要 (torch-tensor > numpy-array)   

Xnp2 = Xtc.detach().numpy()
Xnp2
```

## 2. モデル関数と損失関数の定義

![fig1.png](https://raw.githubusercontent.com/ChemicalBatteryLab-Nitech/dxgem_day4day5/main/Day5_Part2/figs/fig1.png)

```python
model= nn.Sequential(
    nn.Linear(2,1),
    nn.Sigmoid(),
)
```

```python
# modelのパラメータを確認
for p in model.parameters():
    print(p)
```

```python
y=model(X)
y
```

```python
loss_fn = nn.MSELoss()
loss_fn(y,t)  # 予測y と　真値t　間の損失関数(MSE)
```

## 3.モデル関数のWeight と biasの最適化

optimizerの設定  SGD: 性能はよくない<BR>
epoch数、学習率 lr<BR>
model > trainモードにすることで更新<BR>

![fig2.png](https://raw.githubusercontent.com/ChemicalBatteryLab-Nitech/dxgem_day4day5/main/Day5_Part2/figs/fig2.png)

```python
from torch import optim

optimizer = optim.SGD(model.parameters(), lr=0.05)  #model.parameters() 学習させたい変数　　lr は学習率

loss_history=[]
```

```python
model.train()
for epoch in range(10000):  #epoch数 更新処理の回数（ハイパーパラメーター）
    optimizer.zero_grad()   #optimizerの初期化
    y=model(X)
    loss= loss_fn(y,t)  # yとｔでLOSSを評価
    loss_history.append(float(loss))
    loss.backward()   #傾きを計算 (lossを最小化するように)
    optimizer.step()   #更新処理実施

```

```python
y
```

```python
plt.plot(loss_history) # epoch毎のlossの変化をプロット
loss_history[-1]
```
